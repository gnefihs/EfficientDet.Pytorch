```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
mod_inputs = torch.load('inp.pt')

def show_tensor(ori):
    img = ori.squeeze().cpu().permute(1,2,0).detach().numpy()
    img -= np.min(img)
    img /= np.max(img)
    plt.imshow(img[:,:,(2,1,0)])
```

```python
show_tensor(torch.load('x1.pt')[0])
```
