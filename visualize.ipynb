```python
# mod_inputs = torch.load('inp.pt')
def show_tensor(img):
    img = preproc(img)
    plt.figure(figsize = (6,6))
    plt.imshow(img[:,:,(2,1,0)])
#     return img

def preproc(img):
    img = img.squeeze().cpu().permute(1,2,0).detach().numpy()
    img -= np.min(img)
    img /= np.max(img)
    return img
```
